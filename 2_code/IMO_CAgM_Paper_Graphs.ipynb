# Graph 1: Members' number and its evolution over time

```python
#Data dictionary
member_numb_dict = {'1913': 8, '1919': 18, '1921': 29, '1923': 38, '1926': 44, '1929':58, '1932':66, '1935':68, '1937':56, '1946':36, '1947': 48} 
```

```python
# Code for the line plot representing the average attendance of members
import matplotlib.pyplot as plt
from pylab import savefig
import os

years = list(member_numb_dict.keys())
total_members = list(member_numb_dict.values())

plt.plot(years, total_members)
plt.rc('lines', linewidth=4, color='m')
plt.plot(figsize=(10,18))
plt.plot(dpi=600)
plt.title("Members of the IMO technical commission on agriculture")
plt.xlabel("Year")
plt.ylabel("Number of members")
plt.grid(True)
plt.savefig(os.path.join("..","3_plots","Members_lineplot.jpg"), dpi=600)
plt.show()
```

# Graph 3: Heatmap of national participation

```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from pylab import savefig
```

```python
#Data set used to generate the heatmap
df = pd.read_csv(os.path.join("..", "1_data","national_participation_year.csv"), header = 0)
print(df)
```

```python
#The pandas dataframe is reshaped in a numpy array for generating the heatmap
df1 = df.drop(["Nation"], axis =1)
arr = df1.to_numpy()
arr_resh = arr.reshape(37,11)
print(arr_resh)
```

```python
#Code for plotting the heatmap
sns.set_context("paper")
Index = list(df["Nation"])
Cols = ['1913', '1919', '1921', '1923', '1926', '1929', '1932', '1935', '1937', '1946', '1947']
colour = sns.color_palette("coolwarm", 17)
f, ax = plt.subplots(figsize=(11, 15))
ax = sns.heatmap(arr_resh, vmin =0, vmax = 14, cmap= colour, annot=True, fmt="d", linewidths=.5, ax=ax, xticklabels=Cols, yticklabels=Index)
ax.axes.set_title("Heatmap of Attendance per Nation", fontsize=24, y=1.01)
ax.set(xlabel='Year', ylabel='Nation');
savefig(os.path.join("..", "3_plots", 'heatmap.png'), dpi=600)
```
